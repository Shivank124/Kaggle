# Visualization

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(10, 6)) 
india_df.plot(x='Date_reported', y='Cumulative_deaths')
plt.title('Cumulative Deaths in India', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Cumulative Deaths', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(10, 6)) 

# Plot New Deaths
india_df.plot(x='Date_reported', y='New_deaths', label='New Deaths')
plt.title('Deaths in India', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)

plt.legend()
plt.show()
```

It is a quite an intresting plot. We also implement a smooth plot of the data  because the plot does not look smooth enough.

There is two methods here:

    1.Using moving average
    2.Group by

```python
import numpy as np

# Define the window size for the moving average filter
window_size = 7

plt.figure(figsize=(10, 6)) 

# Plot the original data
india_df.plot(x='Date_reported', y='New_deaths', label='New Deaths')

# Apply the moving average filter and plot the smoothed curve
rolling_avg = np.convolve(india_df['New_deaths'], np.ones(window_size)/window_size, mode='same')
plt.plot(india_df['Date_reported'], rolling_avg, label='Smoothed')

plt.title('Deaths in India', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)

plt.legend()
plt.show()
```

Below you can see the smoothed plot

```python
plt.figure(figsize=(10, 6)) 

plt.plot(india_df['Date_reported'], rolling_avg, label='Smoothed')

plt.title('Deaths in India(Smoothed with moving average)', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)

plt.legend()
plt.show()
```

The second methods utlizes groupby. because we used 7 for window in moving average, here we group by week.

```python
# Group the DataFrame by weekly periods based on the 'Date_reported' column
weekly_grouped = india_df.groupby(pd.Grouper(key='Date_reported', freq='W'))

# Compute the sum of the 'New_deaths' column for each week
weekly_sum_deaths = weekly_grouped['New_deaths'].sum()


fig, ax = plt.subplots(figsize=(10, 6))
weekly_sum_deaths.plot(ax=ax, label='New Deaths')

# Set the title and axis labels
ax.set_title('Weekly New Deaths in India', fontsize=18)
ax.set_xlabel('Week', fontsize=14)
ax.set_ylabel('Number of Deaths', fontsize=14)

# Add a legend
ax.legend()

# Display the plot
plt.show()
```

the second method looks better. we implement this plots for Cases too.

```python
plt.figure(figsize=(10, 6)) 

# Plot New Cases
india_df.plot(x='Date_reported', y='New_cases', label='New Cases')

# Plot Cumulative Cases
india_df.plot(x='Date_reported', y='Cumulative_cases', label='Cumulative Cases')

plt.title('Cases in India', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Cases', fontsize=14)

plt.legend()
plt.show()
```

```python
# Group the DataFrame by weekly periods based on the 'Date_reported' column
weekly_grouped = india_df.groupby(pd.Grouper(key='Date_reported', freq='W'))

# Compute the sum of the 'New_deaths' column for each week
weekly_sum_cases = weekly_grouped['New_cases'].sum()


fig, ax = plt.subplots(figsize=(10, 6))
weekly_sum_cases.plot(ax=ax, label='New Cases')

ax.set_title('Weekly New Cases in India', fontsize=18)
ax.set_xlabel('Week', fontsize=14)
ax.set_ylabel('Number of Cases', fontsize=14)
ax.legend()
plt.show()
```

In the below, you can see the new death and new case in one plot.The weekly_sum_deaths series have been made 10 fold bigger for better visualization.

You can also alter the first line to make the death plot bigger or smaller

```python
# Rescale weekly_sum_deaths by 10 fold
weekly_sum_deaths *= 10
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
# Plot the weekly cases with a light blue color
ax.plot(weekly_sum_cases.index, weekly_sum_cases.values, color='#00bfff', label='Weekly Cases')

# Plot the weekly deaths
ax.plot(weekly_sum_deaths.index, weekly_sum_deaths.values, color='lightcoral', label='Weekly Deaths')

# Fill the area under the weekly deaths
ax.fill_between(weekly_sum_deaths.index, weekly_sum_deaths.values, color='red', alpha=0.2)

ax.set_title('Weekly Cases and Deaths', fontsize=18)
ax.set_xlabel('Week', fontsize=14)
ax.set_ylabel('Number of Cases/Deaths', fontsize=14)
ax.legend()
plt.show()
```

# Hypothesis

There is a lag corelatoin between new cases and new death


Also finding the greatest lag corelation

```python
import pandas as pd
import statsmodels.api as sm

# Calculate the cross-correlation between weekly_sum_cases and weekly_sum_deaths
ccf = sm.tsa.stattools.ccf(weekly_sum_cases, weekly_sum_deaths)[0:60]

# Plot the cross-correlation
plt.plot(ccf)
plt.title("Cross-correlation between Weekly Cases and Deaths")
plt.xlabel("Lag")
plt.ylabel("Cross-correlation")
plt.show()
```

It may be because of Groupby that there is not a specific lag correlation in 15 first Week! So we use the daily values.

```python
New_Cases = india_df ["New_cases"]
New_deaths = india_df ["New_deaths"]
```

```python
# Rescale weekly_sum_deaths by 10 fold
New_deaths_copy = 10*New_deaths
New_deaths_copy
```
